### Object-Oriented Programming (OOP) in Python

Object-Oriented Programming is a programming methodology that uses objects and classes to structure code in a more modular and reusable way. Instead of writing functions that operate on data, OOP bundles data and functions together into objects, making code easier to organize, maintain, and scale.

#### Some Terminology

**Class** - A blueprint or template for creating objects. It defines the structure and behavior that instances of that class will have. Think of it like a cookie cutter that defines what shape cookies will be.

**Instance** - A variable that gets assigned a class becomes an instance of that class. It's an actual object created from the class blueprint. If a class is a cookie cutter, an instance is an actual cookie made from that cutter.

**Object** - Another term for an instance. Objects contain both data (attributes) and functions (methods).

#### What is a Class?

A class is a reusable blueprint for creating objects. It defines:
- **Attributes**: Variables that store data about the object (e.g., name, age, price)
- **Methods**: Functions defined inside the class that operate on the object's data

Classes allow you to create multiple instances with the same structure, avoiding code repetition. For example, you might create an `Item` class that can be used to create many different items, each with their own price and quantity.

#### What is a Method?

Methods are functions defined inside classes that act as actions or behaviors for objects. They allow you to perform operations on the object's data.

Key points about methods:
- Always pass at least one argument into a method when you define it: `self`
- `self` represents the instance itself - Python automatically passes the object as the first argument when you call a method
- You access an object's attributes within a method using `self.attribute_name`
- Methods are called using dot notation: `object.method_name()`

Example: `item1.calculate_total_price()` calls the `calculate_total_price` method on the `item1` object.


```python
#Lets define a class named Item 
class Item:
    # to define a veriable to use this class 
    # Object = Item()

    def calculate_total_price(self, price, quantity):
        # to call this method use:
        # Object.calculate_total_price(price, quantity)
        #   In this function:
        #       Object is passed as self
        #       price is passed as  price
        #       quantity is passed as quantity
        return price * quantity

# Lets try it out
item1 = Item() # Creating an instance of class Item
total_price = item1.calculate_total_price(100, 5)
print(total_price)  # Output: 500

# We can also assign item1 variables to store information
# This is known as adding attributes to an object
item1.price = 250
item1.quantity = 8
# Now we can use those variables to calculate the total price
total_price = item1.calculate_total_price(item1.price, item1.quantity)
print(total_price)  # Output: 2000

#It'd be annoying to always have to create and pass item1.price and item1.quantity every time we call calculate_total_price...
```

#### Initailizing Objects
We can initialize Objects to set them up for convienent use with a special method called __init__.

Basically, everytime a new instance of a class is made, init is run automatically after.


```python
#Redefining Class to include initialization
class Item:

    def __init__(self, price, quantity = 0): # Note quantity has a default value of 0
        self.price = price
        self.quantity = quantity # if quantity isn't provided, it defaults to 0

    def calculate_total_price(self): # Note no parameters other than self
        #This method now uses the object's own price and quantity attributes
        #This couldve been done earlier, but it was not garunteed that the object would have those attributes
        #since init creates those attributes, we don't need to explicitly pass them anymore
        return self.price * self.quantity
    
# Lets try it out
item1 = Item(100,5) # Creating an instance of class Item with price and quantity
total_price_item1 = item1.calculate_total_price()
print(total_price_item1)  # Output: 500   

item2 = Item(100) # Creating an instance of class Item with just price 
total_price_item2 = item2.calculate_total_price()
print(total_price_item2)  # Output: 0
```

More on Attritubes: Class vs Instance

So far, we have been assigning our instances attriubes, the variables we've been making like price and quantity. 
Just like instances, the class itself can have attributes which all instances of the class can have. 

```python
#Adding a Class attribute
class Item:
    Discount_Rate = 0.1  # Class Attribute representing a discount rate of 10%
    def __init__(self, price: float, quantity = 0): # Note quantity is set to be an int due it being assigned a default value of 0
        #                ^ define price to be a float (takes ints too)
        #verify variables
        assert price >= 0, "Price must be greater than or equal to zero!"
        assert quantity >= 0, "Quantity must be greater than or equal to zero!"

        # add instance attributes
        self.price = price
        self.quantity = quantity 
        self.total_price = self.calculate_total_price()

    def calculate_total_price(self): 
        return self.price * self.quantity
    
    # New method to apply discount using class attribute
    def apply_discount(self):
        discounted_price = self.total_price * (1 - Item.Discount_Rate)
        return discounted_price
    
# Lets try it out
item1 = Item(200,3) 
print(item1.total_price)  # Output: 600
discounted_price_item1 = item1.apply_discount()
print(discounted_price_item1)  # Output: 540

# What if we want to change the discount rate for a specific item
item1.Discount_Rate = 0.2  # Change discount rate to 20% for only item1
discounted_price_item1 = item1.apply_discount()
print(discounted_price_item1)  # Output: 540 - why?

# Because apply_discount uses Item.Discount_Rate which refers to the class attribute, not the instance attribute.
# To use the instance attribute, we would need to modify apply_discount to use self.Discount_Rate instead.
# self.Discount_Rate would first look for an instance attribute, and if not found, would fall back to the class attribute.
```

How can we see all the Instances of a Class?

```python
#Adding a Class attribute called Instances to keep track of all instances
class Item:
    Instances = []  # Class attribute to keep track of all instances
    def __init__(self, price: float, quantity = 0):
        
        #verify variables
        assert price >= 0, "Price must be greater than or equal to zero!"
        assert quantity >= 0, "Quantity must be greater than or equal to zero!"

        # add instance attributes
        self.price = price
        self.quantity = quantity 
        self.total_price = self.calculate_total_price()

        #add instance to Instances list
        Item.Instances.append(self) #every time an instance is created, it is added to the Instances list

    def calculate_total_price(self): 
        return self.price * self.quantity
    
# Lets try it out
item1 = Item(100,4)
item2 = Item(200,2)
item3 = Item(50,10)
print(Item.Instances)  
# Here we will see it output something like what is below:
# Output: [<__main__.Item object at ...>, <__main__.Item object at ...>, <__main__.Item object at ...>]
# It's telling us the memory addresses of the instances created, whihch is ussually not very informative to the user
```

Lets fix that using a special method - __repr__

explain repr

```python
#Adding __repr__ method to improve instance representation
class Item:
    Instances = []  # Class attribute to keep track of all instances
    def __init__(self, price: float, quantity = 0):
        
        #verify variables
        assert price >= 0, "Price must be greater than or equal to zero!"
        assert quantity >= 0, "Quantity must be greater than or equal to zero!"

        # add instance attributes
        self.price = price
        self.quantity = quantity 
        self.total_price = self.calculate_total_price()

        #add instance to Instances list
        Item.Instances.append(self) #every time an instance is created, it is added to the Instances list

    def __repr__(self):
        # It is good practice to have __repr__ return a string that could be used to recreate the object
        return f"Item({self.price}, {self.quantity})"

    def calculate_total_price(self): 
        return self.price * self.quantity
    
# Lets try it out
item1 = Item(100,4)
item2 = Item(200,2)
item3 = Item(50,10)
print(Item.Instances)  
# Here we will see it output something like what is below:
# Output: [Item(100, 4), Item(200, 2), Item(50, 10)]
# This is much more informative, showing deatails of each item instance.
```
